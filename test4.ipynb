```python

1 C++ Expression Token Counter Using Stringstream   . To count the number of tokens (words/symbols) in a C++ expression using spaces.

#include <iostream>
#include <string>
#include <sstream>
#include <vector>
using namespace std;
int main() {
    string expression;
    cout << "Enter a C++ expression: ";
    getline(cin, expression);

    stringstream ss(expression);
    string token;
    int tokenCount = 0;

    while (ss >> token) {
        tokenCount++;

        }

    cout << "Number of tokens: " << tokenCount << endl;
    return 0;
}











2   CPP Expression Token Counter Using Stringstream

#include <iostream>
#include <string>
#include <sstream>
#include <vector>
using namespace std;
int main() {
    string expression;
    cout << "Enter a C++ expression: " << endl;
    getline(cin, expression);
    stringstream ss(expression);
    string token;
    int tokenCount = 0;
    while (ss >> token) {
        tokenCount++;
        //cout << "Token: " << token << endl;
    }
    cout << "Number of tokens: " << tokenCount << endl;
    return 0;
}












3  CPP  Variable Name Validator

#include <iostream>
#include <string>

using namespace std;

bool isValidVariableName(const string& name) {
    if (name.empty()) {
        return false;
    }

    char firstChar = name[0];
    if (!( (firstChar >= 'a' && firstChar <= 'z') ||
           (firstChar >= 'A' && firstChar <= 'Z') ||
           firstChar == '_')) {
        return false;
    }

    for (size_t i = 1; i < name.length(); ++i) {
        char c = name[i];
        if (!((c >= 'a' && c <= 'z') ||
              (c >= 'A' && c <= 'Z') ||
              (c >= '0' && c <= '9') ||
              c == '_')) {
            return false;
        }
    }

    return true;
}

int main() {
    string variableName;
    cout << "Enter a variable name to check: ";
    cin >> variableName;

    if (isValidVariableName(variableName)) {
        cout << "'" << variableName << "' is a valid Cpp variable name." << endl;
    } else {
        cout << "'" << variableName << "' is NOT a valid Cpp variable name." << endl;
    }

    return 0;
}













4  #include <iostream>
#include <vector>
#include <string>
#include <cctype>
using namespace std;
int main() {
    string input = "if x + 10";
    vector<pair<string, string>> tokens;

    int i = 0;

    while (i < input.length()) {

        if (isspace(input[i])) {
            i++;
            continue;
        }

       if (isalpha(input[i])) {
            string value = "";
            while (i < input.length() && isalpha(input[i])) {
                value += input[i];
                i++;
            }

            if (value == "if") {
                tokens.push_back({"KEYWORD", value});
            } else {
                tokens.push_back({"IDENTIFIER", value});
            }
            continue;
        }

        if (isdigit(input[i])) {
            string value = "";
            while (i < input.length() && isdigit(input[i])) {
                value += input[i];
                i++;
            }
            tokens.push_back({"INTEGER", value});
            continue;
        }

        if (input[i] == '+' || input[i] == '-' || input[i] == '*' || input[i] == '/') {
            string value = string(1, input[i]);
            tokens.push_back({"OPERATOR", value});
            i++;
            continue;
        }
        string value = string(1, input[i]);
        tokens.push_back({"INVALID", value});
        i++;
    }

    for (const auto& token : tokens) {
        cout << "Type: " << token.first << ", Value: " << token.second << endl;
    }

    return 0;
}








5  //C++ Program to Recognize Strings Matching the Pattern a*b+ Using a Simple Finite State Automaton

#include<bits/stdc++.h>
using namespace std;
    int main() {
    int state=0,i,len;
	string s;
	cout << "Enter a String: ";
	cin >> s;
	len=s.size();
	for(i=0; i<len; i++)
    {
		if(state==0 &&s[i]=='a')
        {
            state=0;
        }
        else if(state==0||state==1 &&s[i]=='b')
        {
           state=1;
        }
        else if(state==1&&s[i]=='a')
           {
            state=2;
            break;
           }
        else if(s[i]!= 'a'|| s[i]!='b')
        {
            state=2;
            break;
        }

    }
    if(state==1)
        {cout<<"'"<<s<<"'is accepted under rule 'a*b+'"<<endl;}
    else
        {cout<<"'"<<s<<"'is not recognized."<<endl;}
	return 0;
}


```
